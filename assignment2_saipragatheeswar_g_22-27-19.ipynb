5 Drop all rows with N/A,NA,na values in Num_Bedrooms

```python
dataframe=dataframe[~dataframe['NUM_BEDROOMS'].isin(['na','NaN','N/A'])]
dataframe    
```

6 Replace the NaN and String entries in Num_Bath with previous entry

```python
dataframe['NUM_BATH']=pd.to_numeric(dataframe['NUM_BATH'],errors='coerce')
dataframe['NUM_BATH']=dataframe['NUM_BATH'].fillna(method='ffill')
dataframe
```

7.Replace the empty values in “SQ_FT” with the mean of the all entries.

```python
dataframe['SQ_FT']=dataframe['SQ_FT'].replace('--',np.nan)
dataframe['SQ_FT']=dataframe['SQ_FT'].astype(float)
mean_sqft=dataframe['SQ_FT'].mean()
dataframe['SQ_FT'].fillna(mean_sqft,inplace=True)
dataframe
```

13.Display the state having highest average price per square foot of housing area.

```python
state_prices=dataframe.groupby('STATE')['PRICE','SQ_FT'].mean()
print(state_prices)
state_prices['pricepersqft']=state_prices['PRICE']/state_prices['SQ_FT']
state_prices=state_prices.sort_values('pricepersqft',ascending=True)
print(state_prices.index[0])
```
